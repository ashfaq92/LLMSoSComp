```python
from langchain_mcp_adapters.client import MultiServerMCPClient  
from langchain.agents import create_agent
```

```python
print("Creating Multi-Server MCP Client...")
```

```python
wot_bridge = MultiServerMCPClient(  
    {
        "wot_bridge": {
            "transport": "sse",  # Use Server-Sent Events instead
            "url": "http://localhost:6274/"  # Adjust port based on your server
        }
    }
)
```

```python
tools = await wot_bridge.get_tools()  
```

```python
agent = create_agent(
    "gpt-4o-mini",
    tools  
)
math_response = await agent.ainvoke(
    {"messages": [{"role": "user", "content": "what's (3 + 5) x 12?"}]}
)
weather_response = await agent.ainvoke(
    {"messages": [{"role": "user", "content": "what is the weather in nyc?"}]}
)
```
